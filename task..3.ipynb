Water Jug Problem using DFS (Dynamic)

```python

# Water Jug Problem using DFS (Dynamic & Simple)

def dfs_water_jug(cap_x, cap_y, goal):
    
    stack = [(0, 0, [])]  
    visited = set()

    while stack:
        x, y, path = stack.pop()

        if (x, y) in visited:
            continue

        visited.add((x, y))

        # If goal found
        if x == goal or y == goal:
            print("\n Goal Achieved!\n")
            for step in path:
                print(step)
            print(f"\nFinal State: ({x}, {y})")
            return

        # Apply all possible rules

        # 1. Fill X
        stack.append((cap_x, y, path + [f"Fill X -> ({cap_x}, {y})"]))

        # 2. Fill Y
        stack.append((x, cap_y, path + [f"Fill Y -> ({x}, {cap_y})"]))

        # 3. Empty X
        stack.append((0, y, path + [f"Empty X -> (0, {y})"]))

        # 4. Empty Y
        stack.append((x, 0, path + [f"Empty Y -> ({x}, 0)"]))

        # 5. Pour X -> Y
        transfer = min(x, cap_y - y)
        stack.append((x - transfer, y + transfer,
                      path + [f"Pour X -> Y -> ({x - transfer}, {y + transfer})"]))

        # 6. Pour Y -> X
        transfer = min(y, cap_x - x)
        stack.append((x + transfer, y - transfer,
                      path + [f"Pour Y -> X -> ({x + transfer}, {y - transfer})"]))

    print(" No Solution Found")


# ----------- User Input (Dynamic Part) -----------
x = int(input("Enter capacity of Jug X: "))
y = int(input("Enter capacity of Jug Y: "))
goal = int(input("Enter Goal amount: "))

dfs_water_jug(x, y, goal)
```
